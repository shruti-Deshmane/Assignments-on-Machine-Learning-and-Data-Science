```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix

# Load the data
df = pd.read_excel(r"C:\Users\LENOVO\python datascience basics (hydrabad ds)\ASSIGNMENT DATA SCIENCE\Random Forest\glass.xlsx", sheet_name="glass")
df
```

```python
# Step 1: EDA
# -------------------------
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nData Types:\n", df.dtypes)
print("\nTarget distribution:\n", df["Type"].value_counts())

```

```python

# Step 2: Visualization

sns.pairplot(df, hue="Type")
plt.show()

plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

# Step 3: Preprocessing

X = df.drop("Type", axis=1)
y = df["Type"]

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Use this instead:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)
```

```python

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_bal, y_train_bal)

y_pred_rf = rf.predict(X_test)

# Evaluation
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print("Random Forest (Bagging):")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada = AdaBoostClassifier(n_estimators=100, random_state=42)
ada.fit(X_train_bal, y_train_bal)

y_pred_ada = ada.predict(X_test)

print(" AdaBoost (Boosting):")
print(confusion_matrix(y_test, y_pred_ada))
print(classification_report(y_test, y_pred_ada))
print("Accuracy:", accuracy_score(y_test, y_pred_ada))
```

```python
print("Accuracy Comparison:")
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("AdaBoost Accuracy:", accuracy_score(y_test, y_pred_ada))
```

### Q1. Difference Between Bagging and Boosting

**Bagging**:
- Builds models in parallel.
- Reduces variance.
- Ex: Random Forest

**Boosting**:
- Builds models sequentially.
- Focuses on errors made by previous models.
- Reduces bias and variance.
- Ex: AdaBoost, GradientBoost

### Q2. How to Handle Imbalanced Data

- Use **SMOTE** to oversample minority class.
- Use **class_weight='balanced'**.
- Use **Precision, Recall, F1** instead of Accuracy.
